### What is EDA?
#### Exploratory Data Analysis (EDA) is an approach/philoshophy for data analysis that employsa variety of techniques to :

   * Maximise insight into a dataset;
   * Uncover underlying structure;
   * Extract important variables;
   * Detect outliers and anomalies;
   * Test underlying assumptions;

#### 4. Multivariate Analysis => Understand the relation between several independent variables and the Dependent Variable.
#### 5. Drawing Hypothesis & Inferences => After doing various analysis, we will see whether our Hypothesis hold True or not!

## 3.1. Age

```python
print(df['Age'].describe())
```

```python
tab = pd.crosstab(df.Age, columns = 'Frequency')   #frequency table
tab
```

```python
df.boxplot(column = 'Age',
       grid = False,
       figsize = (6,4),
       )

plt.show()
```

```python
iqr = df.Age.quantile(0.75) - df.Age.quantile(0.25)
print("Interquartile range:",iqr)
ub = df.Age.quantile(0.75) + 1.5*iqr
lb = df.Age.quantile(0.25) - 1.5*iqr

print("upper bound:",ub)
print("lower bound:",lb)
```

```python
below_lb = np.sum(df['Age']<10.5)
print("No. of outliers below lower bound:",below_lb)
above_ub = np.sum(df['Age']>62.5)
print("No. of outliers above upper bound:",above_ub)
total_outliers = above_ub + below_lb
print("Total No. of outliers:",total_outliers)
```

```python
tab.plot(kind='bar', figsize=(12,10))
```

**Inferences**
1. Maximum frequency of people are from age 35 and minimum frequency is of age 60
2. Mean age of the sample is 36.92 ~ 37 years.
3. The data is symmetric in nature.
